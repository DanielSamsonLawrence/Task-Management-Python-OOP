```python
import datetime
```

```python
class Task:
    def __init__(self,title,description,due_date=None,completed=False):
        self.title=title
        self.description=description
        self.created_at=datetime.datetime.now()
        self.due_date=due_date
        self.completed=completed

    def __str__(self):
        status="Completed" if self.completed else "Pending"
        due=f"Due: {self.due_date.strftime('%Y-%m-%d')}" if self.due_date else "No Due Date"
        return f"{self.title}-{status}-{due}"
    
    
```

```python
class TaskManager:
    def __init__(self):
        self.task=[]
        
    def add_task(self, task):
        self.task.append(task)

        
    def view_task(self):
        for i,task in enumerate(self.task,1):
            print(f"{i}.{task}")
            
    
    def update_task(self,index,**kwargs):
        if 1<=index<=len(self.task):
            task=self.task[index-1]
            for key,value in kwargs.items():
                setattr(task,key,value)
            print(f"task {index} updated succesfully")
        else:
            print("Invalid Task")
            
    def delete_task(self,index):
        if 1<=index <= len(self.task):
            del self.task[index-1]
            print(f"Task {index} deleted successfully.")
        else:
            print("Invalid task index.")
```

```python
def main():
    manager = TaskManager()

    while True:
        print("\n1. Add Task")
        print("2. View Tasks")
        print("3. Update Task")
        print("4. Delete Task")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            title = input("Enter task title: ")
            description = input("Enter task description: ")
            due_date_str = input("Enter due date (YYYY-MM-DD) or leave blank: ")
            due_date = datetime.datetime.strptime(due_date_str, "%Y-%m-%d") if due_date_str else None
            task = Task(title, description, due_date)
            manager.add_task(task)
            print("Task added successfully.")

        elif choice == '2':
            manager.view_task()

        elif choice == '3':
            index = int(input("Enter task number to update: "))
            title = input("Enter new title (or leave blank): ")
            description = input("Enter new description (or leave blank): ")
            due_date_str = input("Enter new due date (YYYY-MM-DD) or leave blank: ")
            completed_str = input("Is the task completed? (y/n): ")

            update_dict = {}
            if title:
                update_dict['title'] = title
            if description:
                update_dict['description'] = description
            if due_date_str:
                update_dict['due_date'] = datetime.datetime.strptime(due_date_str, "%Y-%m-%d")
            if completed_str.lower() in ('y', 'yes'):
                update_dict['completed'] = True
            elif completed_str.lower() in ('n', 'no'):
                update_dict['completed'] = False

            manager.update_task(index, **update_dict)

        elif choice == '4':
            index = int(input("Enter task number to delete: "))
            manager.delete_task(index)

        elif choice == '5':
            print("Thank you for using the Task Manager. Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
